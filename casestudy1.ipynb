```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = pd.read_csv("/content/electronics_review.csv")
data
```

```python
# Download NLTK resources
nltk.download("stopwords")
nltk.download("punkt")

# Remove punctuation and stop words, and convert text to lowercase
stop_words = set(stopwords.words("english"))

def preprocess_text(text):
    word_tokens = word_tokenize(text.lower())
    filtered_text = [word for word in word_tokens if word.isalnum() and word not in stop_words]
    return " ".join(filtered_text)

data["Review"] = data["Review"].apply(preprocess_text)

```

```python
# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer(max_features=3000)
tfidf_vectors = tfidf_vectorizer.fit_transform(data["Review"]).toarray()
```

```python
X_train, X_test, y_train, y_test = train_test_split(tfidf_vectors, data["Sentiment"], test_size=0.2, random_state=42)

```

```python
# Random Forest Classifier
classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(X_train, y_train)
```

```python
# Predictions
y_pred = classifier.predict(X_test)

# Model Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

```

```python
def predict_sentiment(text):
    preprocessed_text = preprocess_text(text)
    text_vector = tfidf_vectorizer.transform([preprocessed_text]).toarray()
    sentiment = classifier.predict(text_vector)
    return "Positive" if sentiment == 1 else "Negative"

# Example Usage:
review = "I really love this smartphone. It works perfectly."
print(f"Review: {review}")
print(f"Predicted Sentiment: {predict_sentiment(review)}")
```
